Monitoring tests are supposed to appear at least once a day. A quick heuristic would be to select measurements given at least the same number of times as length of stay.

```python
counts = script_measurement.groupby(['case_number',
                                     'hospital_los_days',
                                     'measurement_concept_code',
                                     'measurement_concept_name']).size().to_frame('measurement_times_per_encounter').reset_index()
```

```python
f = counts['measurement_times_per_encounter'] >= counts['hospital_los_days']
counts.loc[f, ['measurement_concept_code',
               'measurement_concept_name']].drop_duplicates().to_csv('monitoring_tests.csv', index=False)
```

```python
g = counts['measurement_times_per_encounter'] == 1
counts.loc[g, ['measurement_concept_code',
               'measurement_concept_name']].drop_duplicates().to_csv('diagnostic_tests.csv', index=False)
```
