### Dealing with missing data in SOFA scores table

My question: Anna built the SOFA score table for the unified database, and I had noticed that the column corresponding to the total SOFA score for any given day was indeed the sum of all subscores. However, if a subscore was NULL, she would assume it was zero. The problem is that zero in SOFA means the patient did not have failure in that particular organ system, so I asked if I could assume zero whenever data was missing.  
From Cathy Gao: To a certain extent, yes; if a patient is doing very well, they might not have some things (like **bilirubin or PaO2**) checked and other things like GCS might not be documented (though if patients are in the ICU, the nurses are very good at documenting most things).
Most patients should have at least **platelets, MAP, and creatinine** though, so if those are NULL that would be weird.

```python
sofa_with_binary_outcomes.isna().sum()
```

Indeed, the above missing entry count aligns with Cathy's expectations

```python
sofa_with_binary_outcomes = sofa_with_binary_outcomes.dropna(subset=['platelet_points', 'htn_points', 'renal_points'])
```

```python
sofa_with_binary_outcomes.isna().sum()
```

The remaining NULLs will be assumed to be zero, per above

```python
sofa_with_binary_outcomes = sofa_with_binary_outcomes.fillna(0)
```

## Let's do logistic regression on each component of the SOFA score. Rationale: Certain individual SOFA subscores might be more informative on discharge than the overall SOFA score.

The SOFA scores table has daily-calculated scores for each patient encounter/hospitalization, for each day in the ICU (or at least in theory that's how scores are calculated: No more frequent than daily given some measurements that go into the score which are "rate-limiting").  
As such, I'm thinking about three possibilities:  
-Use Day 1 SOFA score: Some literature suggests that day 1 score is good enough for prognosis.  
-Use median SOFA score: To capture a "representative value".  
-Use delta SOFA score: This can be last day SOFA score minus day 1 SOFA score, or (max - min) SOFA score, and  
-(although it seems contrived) The sum of all deltas (from day 1 to day 2, and so on).

Remember that a previous data processing step obviates whether to exclude patients with more than one visit to the ICU during their hospitalization.

### Day 1 SOFA subscores: Some patients will have multiple Day 1 entries since they went in and out of the ICU during a particular hospitalization. Three ways to deal with this:  
1) Just use Day 1 of first ICU visit. that would be the very first SOFA score ever computed.  
2) Take a mean of all those days  
3) Take the worst value (in this case, the max)  

Will stick with option 1 for now

```python
f = (sofa_with_binary_outcomes['icu_rank'] == 1) & (sofa_with_binary_outcomes['ICU_Day'] == 1)
day1_sofa = sofa_with_binary_outcomes.loc[f]
print(f"There are {day1_sofa['case_number'].nunique()} unique patient encounters with a valid Day 1 SOFA score")
```

```python
day1_sofa = day1_sofa.drop(labels=['case_number', 'icu_rank', 'ICU_Day'], axis=1)
```

Logistic regression a la Ritika

```python
from statsmodels.formula.api import logit
```

```python
model_collect = {}
```

Overall initial SOFA

```python
# This formulation (y ~ X) includes the intercept by default. Manually adding an intercept in statsmodels becomes necessary when using sm.Logit(y, X).
overall_initial_SOFA = logit("discharge_disposition_name_conv ~ SOFA", data = day1_sofa).fit()
result_table = flm_tools.logistic_reg_outputter(overall_initial_SOFA, title='Regression: survival ~ initial SOFA')
model_collect['overall_initial_SOFA'] = overall_initial_SOFA
```

```python
# Curious to observe the AUROC for each model
y_true = day1_sofa['discharge_disposition_name_conv']
y_pred = overall_initial_SOFA.predict(day1_sofa['SOFA'])

from sklearn.metrics import roc_curve, auc
fpr, tpr, _ = roc_curve(y_true, y_pred)
auc1 = auc(fpr, tpr)
plt.plot(fpr, tpr, label="AUROC = "+str(round(auc1, 2))+")")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

Initial platelet subscore

```python
initial_platelets = logit("discharge_disposition_name_conv ~ platelet_points", data = day1_sofa).fit()
result_table = flm_tools.logistic_reg_outputter(initial_platelets, title='Regression: survival ~ initial platelet points')
model_collect['initial_platelets'] = initial_platelets
```

```python
y_pred = initial_platelets.predict(day1_sofa['platelet_points'])

from sklearn.metrics import roc_curve, auc
fpr, tpr, _ = roc_curve(y_true, y_pred)
auc1 = auc(fpr, tpr)
plt.plot(fpr, tpr, label="AUROC = "+str(round(auc1, 2))+")")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

Including all initial subscores

```python
subscore_initial_SOFA = logit("discharge_disposition_name_conv ~ P_F_ratio_points + platelet_points + bilirubin_points + htn_points + gcs_points + renal_points", data = day1_sofa).fit()
result_table = flm_tools.logistic_reg_outputter(subscore_initial_SOFA, title='Regression: survival ~ initial SOFA subscores')
model_collect['subscore_initial_SOFA'] = subscore_initial_SOFA
```

```python
y_pred = subscore_initial_SOFA.predict(day1_sofa.drop(labels='discharge_disposition_name_conv', axis=1))

fpr, tpr, _ = roc_curve(y_true, y_pred)
auc1 = auc(fpr, tpr)
plt.plot(fpr, tpr, label="AUROC = "+str(round(auc1, 2))+")")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
df = pd.DataFrame()
names = []
llf = []
llr = []
llrpval = []
aic = []
bic= []
for name, model in model_collect.items():
    # model.save(name, remove_data=True)
    names.append(name)
    llf.append(model.llf)
    llr.append(model.llr)
    llrpval.append(model.llr_pvalue)
    aic.append(model.aic)
    bic.append(model.bic)
    

df['model_name'] = names
df['LL'] = llf
df['LLR'] = llr
df['LLR_pval'] = llrpval
df['AIC'] = aic
df['BIC'] = bic
df
```

Some thoughts:  
1. It's not clear whether PF_ratio is informative. Have to keep in mind that there are data missingness issues.
2. However, it is striking that platelets consistently appeared on both the initial and the median SOFA scores assessments. Considering this is a measurement that should be taken every day, it can either mean that its consistent measurement contributes to its informative value to the models (although not exclusive to platelets in this model), or that indeed it is a very telling measurement. Curiously, it was also an "important" feature when running GradientBoosting.
3. Might be interesting to see clustering just on the platelets. Furthermore, in the idea of using the HPO, there are two types of tests: screening and diagnostic. Platelets is screening. I imagine running regressions on those tests that are of the screening types and ask which ones seem to hold info regarding outcomes. I anticipate glucose featuring heavily. But... I want to consider timing of things!!!

Extra thoughts after discussion with Ritika:  
It is remarkable that the intercept has a comparatively greater absolute value than any of the coefficients, with also a comparatively lower p-value. It's almost as if you'll do fine using averages. Might be good to explore what happens when assessing deviations from baseline score as the actual feature, not so much the actual scores.
