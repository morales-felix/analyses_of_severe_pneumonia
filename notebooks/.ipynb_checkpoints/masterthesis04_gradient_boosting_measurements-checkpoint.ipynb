I'll see whether gradient-boosting (a general-purpose machine learning algorithm) can pick up enough signal in the clinical.measurements table to predict discharge_disposition (outcome) favorability.  

In this notebook, I will use binary outcomes, rather than the 0 to 5 outcomes (multiple outcomes). 
Will also use median value of numerical measurements, therefore restricting features/measurements to those ones. Possible avenues I could take with this:  
1) Include the categorical ones by encoding them.  
2) Only use numerical measurements that showed clear correlations with outcomes.  
3) Transform the analysis to one where I'm only interested in the presence/absence of a measurement.

### Get required tables/data, and manipulate them for use by model

```python
clean_all_table = flm_tools.get_cleaned_data(version='2021-06-02/210602_digest_including_basic_endpoints_and_sofa')

diagnosis = patients.modified_edw_rc('pneumonia_episode_category_assessment', revision='2021-06-02/210602_digest_including_basic_endpoints_and_sofa',
                                     columns=['case_number', 'clin_cap_viral_npop', 'pneu_assess_dt', 'pt_category'])

diagnosis = feature_conversion.shorten_pt_category(diagnosis)
diagnosis = diagnosis.dropna(subset=['pt_category'])
diagnosis['pt_category'] = diagnosis['pt_category'].replace({'VAP': 'HAP/VAP',
                                                            'HAP': 'HAP/VAP'})
diagnosis = diagnosis.drop_duplicates(subset=['case_number'], keep='last')

clean_all_table = pd.merge(clean_all_table, diagnosis)
outcomes = clean_all_table[['case_number', 'discharge_disposition_name_conv']].drop_duplicates()
```

```python
# Preparing tables for GradientBoosting
num_measure_table = clean_all_table[clean_all_table['value_as_number'].notnull()].copy()
num_measure_table['value_as_number'] = num_measure_table['value_as_number'].astype(float)
```

```python
# Obtaining median value
result_median = num_measure_table.groupby(
    ['case_number','discharge_disposition_name_conv',
     'measurement_concept_name'])['value_as_number'].median().to_frame('median_value').reset_index()
```

#### Note that this is a detour: Clustermaps won't show without me taking out measurements which have a null Spearman correlation p-value.

```python
# This line is to take away measurements appearing only once
c = result_median['measurement_concept_name'].value_counts()
result_median = result_median[result_median['measurement_concept_name'].isin(c[c>1].index)]

# Making the measurements the row indices.
result_median = result_median.set_index('measurement_concept_name')

agg_results = []

for test in result_median.index.unique():
    discharges = result_median.loc[test, 'discharge_disposition_name_conv']
    measurement_values = result_median.loc[test, 'median_value']
    
    rho, pvalue = stats.spearmanr(discharges, measurement_values)
    results = {
        'Measurement': test,
        'Corr_values_Spearman': rho,
        'P_values_Spearman': pvalue,
        'Bonferroni_pvalue': np.clip([pvalue*len(result_median.index.unique())], 0, 1)[0],
        '#_cases_given_test': len(discharges)
    }
    agg_results.append(results)
    
results = pd.DataFrame(agg_results)
results_copy = results.dropna(subset=['Corr_values_Spearman','P_values_Spearman', 'Bonferroni_pvalue']).copy()
significant_features = list(results_copy[(results_copy.Bonferroni_pvalue < 0.01) & (results_copy.Bonferroni_pvalue > 0)]['Measurement'])

with open("Spearman_tests.txt", "w") as f:
    for s in significant_features:
        f.write(str(s) + "\n")
```

```python
working_table = clean_all_table[['case_number','measurement_concept_name',
                                   'discharge_disposition_name_conv']].drop_duplicates()

agg = []

# Create a pandas series keyed by case_number, returning the specific outcome of that case
outcome_by_patient = outcomes.set_index('case_number', verify_integrity=True)['discharge_disposition_name_conv']

# For each measurement in the table
for test in working_table['measurement_concept_name'].unique():
    
    # First, get an array of True/False to flag rows containing the test we'd like to analyze
    f = working_table['measurement_concept_name'] == test
    
    # Now getting the case_numbers (patients) who got the test into a list. This is necessary for later.
    posit_ids = list(working_table.loc[f,'case_number'])
    
    # Now gather that outcome of each of the patients who got the test
    posit_out = outcome_by_patient[posit_ids]

    # And the outcome of those who didn't get the test
    negat_out = outcome_by_patient[~outcome_by_patient.index.isin(posit_ids)]
    
    # Finally, evaluate Mann-Whitney on both lists
    _, pvaluemw = stats.mannwhitneyu(posit_out,negat_out,alternative='two-sided')
        
    # And gather results in a row of a table
    results1={
        'Measurement': test,
#         'U_values_MannWhitney': u,
        'p_values_MannWhitney': pvaluemw,
        'bonferroni_pvalues': np.clip([pvaluemw*len(working_table['measurement_concept_name'].unique())], 0, 1)[0],
        'cases_given_test': len(posit_out),
        'Median_present': np.median(posit_out),
        'Median_absent': np.median(negat_out),
#         'IQR_present': stats.iqr(posit_out),
#         'IQR_absent': stats.iqr(negat_out),
#         'cases_NOT_given_test': len(negat_out)
    }
    agg.append(results1)
    
second_results = pd.DataFrame(agg)
second_results['Median_diff'] = second_results['Median_present'] - second_results['Median_absent']

filtered_MannWhitney = second_results.dropna()
filtered_MannWhitney_sig = filtered_MannWhitney.loc[filtered_MannWhitney.bonferroni_pvalues < 0.01]
associated_measurements = list(filtered_MannWhitney_sig['Measurement'])

with open("Mann_Whitney_tests.txt", "w") as f:
    for s in associated_measurements:
        f.write(str(s) + "\n")
```

### Gradient Boosting Classifier!

```python
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
%%time
store_feature_importances_of_individual_iterations= []
AUCs = []
accuracies = []
precisions = []
recalls = []

X = da_DF.drop(output[0], axis='columns')
y = da_DF.loc[:,output[0]]
number_of_cases = len(y)
print(number_of_cases)

# Here, because we have few data and want to train from many records still: use "Monte-Carlo cross validation": 
# use one part of data for training, other for testing, and repeat so randomly multiple times
for iteration in range(0, number_of_cases):
    
    # Use 95% of data for training and 5% for testing (because of small overall size of dataset)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05,
        random_state=iteration)    # ensuring that randomly chosen numbers reproduce across multiple executions of this notebook, but are still different random numbers for different iterations of the model
    
    # Instantiate the model
    gbc = GradientBoostingClassifier(loss='deviance', random_state=iteration)
    
    # Fit training data (features and value to predict) to model
    gbc.fit(X_train, y_train)
    
    # Prediction based on features held out for testing/validating
    y_pred = gbc.predict_proba(X_test)
    
    # Calculating various classification performance scores
    AUCs.append(roc_auc_score(y_test,y_pred))
    accuracies.append(accuracy_score(y_test,y_pred))
    precisions.append(precision_score(y_test,y_pred))
    recalls.append(recall_score(y_test,y_pred))
    
    # Prepare output of this round of iteration. We want to keep track of three things:
    # - the original values for readout_of_interest
    # - the predicted values corresponding to these original values
    # - Likely not required, but could be helpful for troubleshooting in case of unexpected results: the number given to current iteration   
    out = pd.DataFrame({
        'feature': X.columns,       # name of feature
        'importance': gbc.feature_importances_,    # importance of feature toward model
        'iteration': iteration
    })
    store_feature_importances_of_individual_iterations.append(out)
```

```python
CV_results = [AUCs, accuracies, precisions, recalls]
for_labels = ['AUC', 'Accuracy', 'Precision', 'Recall']

fig1, ax1 = plt.subplots(1,1)
ax1.boxplot(CV_results, notch='True', bootstrap=10000, patch_artist=True, labels=for_labels)

ax1.set_ylabel("Values of classification scores",fontsize=13)
ax1.grid(linestyle=':')

# plt.savefig("GBC performance.png")
plt.show()
```

```python
all_feature_importances = pd.concat(store_feature_importances_of_individual_iterations)
feature_importances_over_iterations = all_feature_importances.pivot(index='feature', columns='iteration', values='importance')
```

```python
# cluster feature importances over different iterations, if there are row-wise lines it means that same features have similar contribution to different instances of the model
sns.clustermap(feature_importances_over_iterations, cmap='bwr', method='ward', figsize=(18,12))
# plt.savefig("feature_importance.png")
plt.show()
```
