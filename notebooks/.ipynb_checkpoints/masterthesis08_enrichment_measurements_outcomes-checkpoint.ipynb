```python
clean_all_table = flm_tools.get_cleaned_data()

diagnosis = patients.modified_edw_rc('pneumonia_episode_category_assessment', revision='latest',
                                     columns=['case_number', 'clin_cap_viral_npop', 'pneu_assess_dt', 'pt_category'])

diagnosis = feature_conversion.shorten_pt_category(diagnosis)
diagnosis = diagnosis.dropna(subset=['pt_category'])
diagnosis['pt_category'] = diagnosis['pt_category'].replace({'VAP': 'HAP/VAP',
                                                            'HAP': 'HAP/VAP'})
diagnosis = diagnosis.drop_duplicates(subset=['case_number'], keep='last')

clean_all_table = pd.merge(clean_all_table, diagnosis)

measurements_in_encounter = clean_all_table[['case_number','measurement_concept_name',
                                   'discharge_disposition_name_conv']].drop_duplicates()
outcomes = clean_all_table[['case_number', 'discharge_disposition_name_conv']].drop_duplicates()
total_cases = measurements_in_encounter['case_number'].unique()
```

```python
working_table = measurements_in_encounter.copy()
agg = []

# Create a pandas series keyed by case_number, returning the specific outcome of that case
outcome_by_patient = outcomes.set_index('case_number', verify_integrity=True)['discharge_disposition_name_conv']

# For each measurement in the table
for test in working_table['measurement_concept_name'].unique():
    
    # First, get an array of True/False to flag rows containing the test we'd like to analyze
    f = working_table['measurement_concept_name'] == test
    
    # Now getting the case_numbers (patients) who got the test into a list. This is necessary for later.
    posit_ids = list(working_table.loc[f,'case_number'])
    
    # Now gather that outcome of each of the patients who got the test
    posit_out = outcome_by_patient[posit_ids]

    # And the outcome of those who didn't get the test
    negat_out = outcome_by_patient[~outcome_by_patient.index.isin(posit_ids)]
    
    # Finally, evaluate Mann-Whitney on both lists
    _, pvaluemw = stats.mannwhitneyu(posit_out,negat_out,alternative='two-sided')
        
    # And gather results in a row of a table
    results1={
        'Measurement': test,
        'p_values_MannWhitney': pvaluemw,
        'bonferroni_pvalues': np.clip([pvaluemw*len(working_table['measurement_concept_name'].unique())], 0, 1)[0],
        'cases_given_test': len(posit_out),
        'fraction_patients': len(posit_out)/len(total_cases),
        'Median_present': np.median(posit_out),
        'Median_absent': np.median(negat_out),
    }
    agg.append(results1)
    
second_results = pd.DataFrame(agg)
second_results['Median_diff'] = second_results['Median_present'] - second_results['Median_absent']
```

```python
filtered_MannWhitney = second_results.dropna()
filtered_MannWhitney.loc[:,'log_bonferroni'] = -1*np.log10(filtered_MannWhitney['bonferroni_pvalues'])
filtered_MannWhitney_sig = filtered_MannWhitney.loc[filtered_MannWhitney.bonferroni_pvalues < 0.01]
filtered_MannWhitney_nonsig = filtered_MannWhitney.loc[filtered_MannWhitney.bonferroni_pvalues >= 0.01]
```

```python
text1 = ["aPPT in Platelet poor plasma", "Calcium.ionized in Blood",
         "Fibrinogen in Platelet poor plasma", "Burr cells in Blood",
        "Haptoglobin in Serum"]
```

```python
a = results.copy()

f = a['log_bonferroni'] >= 2
f_increase = a['ratio'] > 0
```

```python
for j in a[f].index:
    if a.loc[j, 'measurement_concept_name'] == 'Amikacin [Mass/volume] in Serum or Plasma':
        a.loc[j, 'measurement_concept_name'] = 'Amikacin in Serum'
    elif a.loc[j, 'measurement_concept_name'] == 'Burr cells [Presence] in Blood by Light microscopy':
        a.loc[j, 'measurement_concept_name'] = 'Burr cells in Blood'
    else:
        a.loc[j, 'measurement_concept_name'] = 'Fibrinogen in Platelet poor plasma'
```

```python
fig5,ax5 = plt.subplots(ncols=2, nrows=1, figsize=(16,7))
ax5[0].scatter(filtered_MannWhitney_sig['Median_diff'], filtered_MannWhitney_sig['log_bonferroni'],
            marker='o', s=filtered_MannWhitney_sig['cases_given_test']/5, c='r')
ax5[0].scatter(filtered_MannWhitney_nonsig['Median_diff'], filtered_MannWhitney_nonsig['log_bonferroni'],
            marker='o', s=filtered_MannWhitney_nonsig['cases_given_test']/5, c='darkgray')
ax5[0].set_xlabel("Median outcome difference", fontsize=20)
ax5[0].set_ylabel(r"$-log_{10}$(Adjusted p-value)", fontsize=20)
ax5[0].tick_params(axis='x', labelsize=20)
ax5[0].tick_params(axis='y', labelsize=20)
ax5[0].vlines(0, 0, ax5[0].get_ylim()[-1], color='lightgrey', linestyles='dashed')
ax5[0].grid(linestyle=':')

x = list(filtered_MannWhitney_sig['Median_diff'])
y = list(filtered_MannWhitney_sig['log_bonferroni'])

for i in range(0,len(x)):
    if text1[i] == "Fibrinogen in Platelet poor plasma":
        ax5[0].annotate(text1[i], (x[i]-1, y[i]-0.4), fontsize=15)
    elif text1[i] == "Calcium.ionized in Blood":
        ax5[0].annotate(text1[i], (x[i]+0.1, y[i]), fontsize=15)
    elif text1[i] == "Haptoglobin in Serum":
        ax5[0].annotate(text1[i], (x[i]-1.3, y[i]+0.4), fontsize=15)
    elif text1[i] == "Burr cells in Blood":
        ax5[0].annotate(text1[i], (x[i]-0.2, y[i]-0.4), fontsize=15)
    else:
        ax5[0].annotate(text1[i], (x[i]-0.9, y[i]+0.2), fontsize=15)
        

ax5[1].scatter(np.log10(a.loc[~f, 'dead_in_measurement'] + 1),
               np.log10(a.loc[~f, 'survived_in_measurement'] + 1),
               c='darkgray')
ax5[1].scatter(np.log10(a.loc[f & f_increase, 'dead_in_measurement'] + 1),
               np.log10(a.loc[f & f_increase, 'survived_in_measurement'] + 1),
               c='blue')
ax5[1].scatter(np.log10(a.loc[f & ~f_increase, 'dead_in_measurement'] + 1),
               np.log10(a.loc[f & ~f_increase, 'survived_in_measurement'] + 1),
               c='r')

for j in a[f].index:
    ax5[1].text(np.log10(a.loc[j, 'dead_in_measurement']+1)-0.15,
             np.log10(a.loc[j, 'survived_in_measurement']+1)-0.1,
             a.loc[j, 'measurement_concept_name'],
             fontsize=15)
ax5[1].set_xlabel(r"Expired in assay ($log_{10}$)", fontsize=20)
ax5[1].set_ylabel(r"Survived in assay ($log_{10}$)", fontsize=20)
ax5[1].tick_params(axis='x', labelsize=20)
ax5[1].tick_params(axis='y', labelsize=20)
ax5[1].grid(linestyle=':')

plt.tight_layout()
plt.savefig('test_presence_absence_210602.png')
```

```python
clean_all_table = flm_tools.get_cleaned_data(outcome_encoding='binary')

diagnosis = patients.modified_edw_rc('pneumonia_episode_category_assessment', revision='latest',
                                     columns=['case_number', 'clin_cap_viral_npop', 'pneu_assess_dt', 'pt_category'])

diagnosis = feature_conversion.shorten_pt_category(diagnosis)
diagnosis = diagnosis.dropna(subset=['pt_category'])
diagnosis['pt_category'] = diagnosis['pt_category'].replace({'VAP': 'HAP/VAP',
                                                            'HAP': 'HAP/VAP'})
diagnosis = diagnosis.drop_duplicates(subset=['case_number'], keep='last')

clean_all_table = pd.merge(clean_all_table, diagnosis)
clean_all_table = flm_tools.add_column_with_day_of_stay(clean_all_table)

# In case some measurements are encoded in SNOMED, restrict measurements to those encoded in LOINC
f = clean_all_table['measurement_vocabulary_id'] == 'LOINC'
clean_all_table = clean_all_table.loc[f]
```

```python
value = lohpo.AnnotationLoader.load()
header, result = lohpo.HpoTermListLoader.load()
```

```python
codes = set(clean_all_table["measurement_concept_code"])

frames = []
    
for code in codes:
    f = clean_all_table['measurement_concept_code'] == code
    df = clean_all_table.loc[f]
    loincId = lohpo.LoincId.from_code(code)
    loinc2hpo = value.get(loincId, None)
    
    if loinc2hpo is not None:
        for hpo_term in loinc2hpo.candidateHpoTerms.values():
            if hpo_term.isNegated:
                df.loc[:, 'negated_HPO_code'] = hpo_term.id
                hpo_annot = result.get(hpo_term.id)
                df.loc[:, 'HPO_annotation'] = hpo_annot.name
        frames.append(df)
    else:
        df.loc[:, 'negated_HPO_code'] = None
        df.loc[:, 'HPO_annotation'] = None
        frames.append(df)
        
clean_all_table = pd.concat(frames)
```

```python
h = clean_all_table['HPO_annotation'].notnull()
complete_HPO = clean_all_table.loc[h]

outcomes = clean_all_table[['case_number', 'discharge_disposition_name_conv']].drop_duplicates()
outcomes['discharge_disposition_name_conv'] = outcomes['discharge_disposition_name_conv'].astype(bool)
```

```python
hpo_enrich = enrichment(outcomes, complete_HPO, 'case_number', 'discharge_disposition_name_conv', 'HPO_annotation')
```
