```python
dataset_version = '2021-08-31/210831_digest'
sig_thresh = 0.01
```

#### Let's do it for pneumonia types, which measurements are more prevalent in each pneumonia type, as opposed to control

```python
clean_all_table = flm_tools.get_cleaned_data(version=dataset_version, outcome_encoding='binary')
```

#### First, all pneumonias vs. control

```python
diagnosis = patients.modified_edw_rc('pneumonia_episode_category_assessment', revision=dataset_version,
                                     columns=['case_number', 'pt_category'])

diagnosis = feature_conversion.shorten_pt_category(diagnosis)

diagnosis['pt_category'] = diagnosis['pt_category'].replace({
    'HAP': True,
    'CAP': True,
    'VAP': True,
    'non': False})

diagnosis = diagnosis.dropna(subset=['pt_category'])

# Some categories are simply reassessed and reconfirmed. Dropping those duplicates
# Using last, as above pneumonia transformation yields all transitions between categories towards pneumonia.
# So if final category was pneumonia, take that as patient's actual pneumonia category.
# Also, no instance of final category being non.
# There are CAP -> Null transitions, but that's considered pneumonia.
diagnosis = diagnosis.drop_duplicates(subset=['case_number'], keep='last')

table_with_diagnoses = pd.merge(clean_all_table, diagnosis)
table_with_diagnoses = table_with_diagnoses[['case_number', 'pt_category']].drop_duplicates()
```

```python
len(table_with_diagnoses['case_number'].drop_duplicates())
```

```python
category_in_annotation = ['pneumonia_in_measurement', 'control_in_measurement']
category_not_in_annotation = ['pneumonia_not_in_measurement', 'control_not_in_measurement']

results = flm_tools.enrichment(table_with_diagnoses, clean_all_table, 'case_number', 'pt_category', 'measurement_concept_name', category_in_annotation, category_not_in_annotation)
```

```python
flm_tools.plot_step(results, category_in_annotation, 'Pneumonia vs. non-pneumonia', sig_thresh, namer='measurement_concept_name')
```

```python
results.sort_values(by='log_bonferroni', ascending=False)
```

#### HAP/VAP vs. control

```python
diagnosis = patients.modified_edw_rc('pneumonia_episode_category_assessment', revision=dataset_version,
                                     columns=['case_number', 'pt_category'])

diagnosis = feature_conversion.shorten_pt_category(diagnosis)

# VAP is a subset of HAP
diagnosis['pt_category'] = diagnosis['pt_category'].replace({
    'HAP': True,
    'CAP': None,
    'VAP': True,
    'non': False})

diagnosis = diagnosis.dropna(subset=['pt_category'])

# Some categories are simply reassessed and reconfirmed. Dropping those duplicates
# Using last, as above HAP/VAP transformation yields all transitions between categories towards HAP.
# So if final category was HAP, take that as patient's actual pneumonia category.
# Also, no instance of final category being CAP.
diagnosis = diagnosis.drop_duplicates(subset=['case_number'], keep='last')

table_with_diagnoses = pd.merge(clean_all_table, diagnosis)
table_with_diagnoses = table_with_diagnoses[['case_number', 'pt_category']].drop_duplicates()
```

```python
category_in_annotation = ['HAP/VAP_in_measurement', 'control_in_measurement']
category_not_in_annotation = ['HAP/VAP_not_in_measurement', 'control_not_in_measurement']

results = flm_tools.enrichment(table_with_diagnoses, clean_all_table, 'case_number', 'pt_category', 'measurement_concept_name', category_in_annotation, category_not_in_annotation)
```

```python
flm_tools.plot_step(results, category_in_annotation, 'HAP/VAP vs. non-pneumonia', sig_thresh, namer='measurement_concept_name')
```

```python
results.sort_values(by='log_bonferroni', ascending=False)
```

#### CAP vs. control

```python
diagnosis = patients.modified_edw_rc('pneumonia_episode_category_assessment', revision=dataset_version,
                                     columns=['case_number', 'pt_category'])

diagnosis = feature_conversion.shorten_pt_category(diagnosis)

# CAP without transitioning to HAP/VAP
c = diagnosis['case_number'].value_counts()
diagnosis = diagnosis[diagnosis['case_number'].isin(c[c==1].index)]

diagnosis['pt_category'] = diagnosis['pt_category'].replace({'HAP': None,
                                                             'CAP': True,
                                                             'VAP': None,
                                                             'non': False})

diagnosis = diagnosis.dropna(subset=['pt_category'])

table_with_diagnoses = pd.merge(clean_all_table, diagnosis)
table_with_diagnoses = table_with_diagnoses[['case_number', 'pt_category']].drop_duplicates()
```

```python
category_in_annotation = ['CAP_in_measurement', 'control_in_measurement']
category_not_in_annotation = ['CAP_not_in_measurement', 'control_not_in_measurement']

results = flm_tools.enrichment(table_with_diagnoses, clean_all_table, 'case_number', 'pt_category', 'measurement_concept_name', category_in_annotation, category_not_in_annotation)
```

```python
flm_tools.plot_step(results, category_in_annotation, 'CAP vs. non-pneumonia', sig_thresh, namer='measurement_concept_name')
```

```python
results.sort_values(by='log_bonferroni', ascending=False)
```

#### HAP vs. CAP

```python
diagnosis = patients.modified_edw_rc('pneumonia_episode_category_assessment', revision=dataset_version,
                                     columns=['case_number', 'pt_category'])

diagnosis = feature_conversion.shorten_pt_category(diagnosis)

# CAP without transitioning to HAP/VAP
c = diagnosis['case_number'].value_counts()
diagnosis = diagnosis[diagnosis['case_number'].isin(c[c==1].index)]

diagnosis['pt_category'] = diagnosis['pt_category'].replace({'VAP': 'HAP/VAP',
                                                            'HAP': 'HAP/VAP'})

diagnosis['pt_category'] = diagnosis['pt_category'].replace({'HAP/VAP': True,
                                                             'CAP': False,
                                                             'non': None})

diagnosis = diagnosis.dropna(subset=['pt_category'])

diagnosis = diagnosis.dropna(subset=['pt_category'])
table_with_diagnoses = pd.merge(clean_all_table, diagnosis)
table_with_diagnoses = table_with_diagnoses[['case_number', 'pt_category']].drop_duplicates()
```

```python
category_in_annotation = ['HAP/VAP_in_measurement', 'CAP_in_measurement']
category_not_in_annotation = ['HAP/VAP_not_in_measurement', 'CAP_not_in_measurement']

results = flm_tools.enrichment(table_with_diagnoses, clean_all_table, 'case_number', 'pt_category', 'measurement_concept_name', category_in_annotation, category_not_in_annotation)
```

```python
flm_tools.plot_step(results, category_in_annotation, 'HAP/VAP vs. CAP', sig_thresh, namer='measurement_concept_name')
```

```python
results.sort_values(by='log_bonferroni', ascending=False)
```
