```python
dataset_version = '2021-08-31/210831_digest'
sig_thresh = 0.01
```

#### Loading up data tables

```python
clean_all_table = flm_tools.get_cleaned_data(version=dataset_version, outcome_encoding='binary')
```

```python
map_patient_ids = clean_all_table[['patient_ir_id','cohort_patient_id','case_number']].drop_duplicates()      # Needed below

outcomes = clean_all_table[['case_number', 'discharge_disposition_name_conv']].drop_duplicates()
outcomes['discharge_disposition_name_conv'] = outcomes['discharge_disposition_name_conv'].astype(bool)
```

```python
comorb = patients.modified_edw_rc('patient_comorbidities', revision=dataset_version, columns=['patient_ir_id', 'comorbidity'])
```

```python
# Some comorbidity annotations are redundant. To a non-clinician, redundancy in HIV AIDS is obvious
# But others not so much. Would be good to double-check list.

comorb_merge = pd.merge(map_patient_ids, comorb)      # Here, this is why map_patient_ids is needed
comorb_merge = comorb_merge.replace("AIDS HIV", "HIV_AIDS")
comorb_merge = comorb_merge.replace("AIDS/H1V", "HIV_AIDS")
```

```python
comorb_merge = comorb_merge[['case_number', 'comorbidity']].drop_duplicates()
comorb_merge['comorbidity'].drop_duplicates()
```

#### Alright, let's see which comorbidities are enriched in expired vs. survived patients.

```python
category_in_annotation = ['survived_had_comorbidity', 'died_had_comorbidity']
category_not_in_annotation = ['survived_didnot_have_comorbidity', 'died_didnot_have_comorbidity']

results = flm_tools.enrichment(outcomes, comorb_merge, 'case_number', 'discharge_disposition_name_conv', 'comorbidity', category_in_annotation, category_not_in_annotation)
results.sort_values(by='log_bonferroni', ascending=False)
```

```python
flm_tools.plot_step(results, category_in_annotation, 'Survived vs. Expired', sig_thresh, namer='comorbidity')
```
