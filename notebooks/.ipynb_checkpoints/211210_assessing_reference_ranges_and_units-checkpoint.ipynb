### Now, let's add the HPO mapping

```python
# Loading a dictionary containing the mapped LOINC to HPO codes
value = lohpo.AnnotationLoader.load()

# Loading a dictionary containing the HPO terms associated with the HPO codes
header, result = lohpo.HpoTermListLoader.load()
```

```python
# This is the code block that converts measurements in dataset into HPO annotations
# At this point, simply the general physiological annotation (abnormal lymphocyte count,
# i.e. it doesn't specify result of test, just which variable the test assessed)

codes = set(data_december["measurement_concept_code"])

frames = []

for code in codes:
    f = data_december['measurement_concept_code'] == code
    df = data_december.loc[f].copy()
    loincId = lohpo.LoincId.from_code(code)
    loinc2hpo = value.get(loincId, None)
    
    if loinc2hpo is not None:
        for hpo_term in loinc2hpo.candidateHpoTerms.values():
            if hpo_term.isNegated:
                df.loc[:, 'general_HPO_code'] = str(hpo_term.id)
                hpo_annot = result.get(hpo_term.id)
                df.loc[:, 'general_HPO_annotation'] = str(hpo_annot.name)
                
        frames.append(df)
    else:
        df.loc[:, 'general_HPO_code'] = None
        df.loc[:, 'general_HPO_annotation'] = None
        frames.append(df)
data_december = pd.concat(frames)
data_december = data_december.sort_values(by=['case_number',
                                              'day_of_visit',
                                              'order_of_measurement'])
```

#### There are instances where a value for reference range is present in `reference_normal` column, but nothing is present in `reference_low` and `reference_high`. What follows is the parsing of entries in `reference_normal` to then fill out `reference_low` and `reference_high`. See: https://github.com/NUSCRIPT/script_etl_eda/issues/199

```python
mask = data_final['reference_low'].isna() & data_final['reference_high'].isna() & data_final['reference_normal'].notnull()
mask.sum()
```
